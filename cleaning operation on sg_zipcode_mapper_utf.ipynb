```python
# duplicated geocode "BLK 216 AND 215 CHOA CHU KANG CENTRAL"
# Add one data row for BLK 215

hdb_locations = hdb_locations[~hdb_locations['building'].str.contains('BLK 216 AND')] 

newrow = {'postal': '680216', 'lat': '1.3831978', 'lng': '103.7471233', 'searchval': 'HDB-CHOA CHU KANG',
          'block': '216','street_name': 'CHOA CHU KANG CTRL', 'building': 'HDB-CHOA CHU KANG',
          'address': '216 CHOA CHU KANG CTRL HDB-CHOA CHU KANG SINGAPORE 680216', 'postal.1': '680216'}

new_df = pd.DataFrame([newrow])

# Concatenate with the original DataFrame
hdb_locations = pd.concat([hdb_locations, new_df], ignore_index=True)
```

```python
#### Pattern Matching Cleaning on hdb_locations ####
####################################################

# Replace 'AVENUE' with 'AVE' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('AVENUE', 'AVE')

# Replace 'NORTH' with 'NTH' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('NORTH', 'NTH')

# Replace 'ROAD' with 'RD' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('ROAD', 'RD')

# Replace 'STREET' with 'ST' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('STREET', 'ST')

# Replace 'JALAN' with 'JLN' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('JALAN', 'JLN')

# Replace 'PLACE' with 'PL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('PLACE', 'PL')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('CLOSE', 'CL')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('TANJONG', 'TG')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('MARKET', 'MKT')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('LORONG', 'LOR')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('CENTRAL', 'CTRL')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('BUKIT', 'BT')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('UPPER', 'UPP')
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('UPPERER', 'UPP')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('KAMPONG', 'KG')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('SAINT', 'ST.')

# Replace 'CLOSE' with 'CL' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('PARK', 'PK')

# Replace 'DRIVE' with 'DR' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('DRIVE', 'DR')

# Replace 'DRIVR' with 'DR' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('DRIVE', 'DR')

# Replace 'DRIVR' with 'DR' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('NTHSHORE', 'NORTHSHORE')

# Replace 'DRIVR' with 'DR' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('COMMONWEALTH', "C'WEALTH")

# Replace 'DRIVR' with 'DR' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('CRESCENT', "CRES")

# Replace 'DRIVR' with 'DR' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('GARDENS', "GDNS")

# Replace 'DRIVR' with 'DR' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('SOUTH', "STH")

# Replace 'DRIVR' with 'DR' in the 'street_name' column
hdb_locations['street_name'] = hdb_locations['street_name'].str.replace('TERRACE', "TER")
```

```python
# Identify any duplicated block+streetname codes
duplicate_rows = hdb_locations[hdb_locations.duplicated(subset=['block', 'street_name'], keep=False)]
duplicate_rows
```

```python
# Identify any duplicated geo codes
duplicate_geocode = hdb_locations[hdb_locations.duplicated(subset=['postal','lat', 'lng'], keep=False)]
duplicate_geocode
```

```python
# Drop duplicated geolocations
hdb_locations = hdb_locations.drop_duplicates(subset=['block', 'street_name'])
hdb_locations = hdb_locations.drop_duplicates(subset=['postal.1','lat', 'lng'])
```

```python
# Identify any duplicated postal codes
duplicate_postal = hdb_locations[hdb_locations.duplicated(subset=['postal.1'], keep=False)]
duplicate_postal
```

```python
# Clean HDB locations dataset
cleaned_hdb_locations = hdb_locations.drop(['postal','address', 'searchval','building'], axis = 1)
cleaned_hdb_locations = cleaned_hdb_locations.drop_duplicates()

#Rename 'postal.1' column
cleaned_hdb_locations.rename(columns={'postal.1':'postal'},inplace=True)
```

```python
cleaned_hdb_locations = cleaned_hdb_locations.reset_index(drop=True)
cleaned_hdb_locations
```
