```python
import pandas as pd
```

```python
hawkers = pd.read_csv('ListofGovernmentMarketsHawkerCentres.csv')
```

```python
hawkers.head()
```

```python
geocode = pd.read_csv('updated_sg_zipcode_mapper_utfv2.csv', index_col = 0)
geocode.head()
```

```python
# Removal of 'Blk' for easier API extraction
hawkers['name_of_centre'] = hawkers['name_of_centre'].str.replace('Blk ','')
```

```python
hawkers_dataset = pd.merge(hawkers, geocode, how='left', left_on='name_of_centre',right_on='searchval')
hawkers_dataset.head()
```

```python
mismatch_hawkers = hawkers_dataset[hawkers_dataset['lat'].isna()]
mismatch_hawkers

## OBSERVATION: '4A Woodlands Centre Road' has been closed down.
```

```python
hawkers_dataset.drop(62, inplace=True)
```

```python
hawkers_dataset = hawkers_dataset[['name_of_centre','type_of_centre','postal','lat','lng','block','street_name']]
hawkers_dataset.reset_index(drop=True)
```

```python
hawkers_dataset.to_csv('hawkers_dataset.csv', index = False)
```
